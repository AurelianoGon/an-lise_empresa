## 2. Análise de tendência das variáveis

Os dados estão distribuídos no meses, podemos analisar a evolução das variáveis mês a mês.

```python
# Definir o estilo dos gráficos
sns.set(style="whitegrid")

# Criar uma figura e eixos
plt.figure(figsize=(14, 8))

# Plotar a tendência de custo_marketing
sns.lineplot(x='mes', y='custo_marketing', data=df, marker='o', 
             label='Custo Marketing', color='blue')

sns.lineplot(x='mes', y='custo_produtos', data=df, marker='o',
             label='Custo Produtos', color='red')

sns.lineplot(x='mes', y='custo_entrega', data=df, marker='o', 
             label='Custo Entrega', color='green')

plt.title('Tendência dos Custos ao Longo dos Meses')
plt.xlabel('Meses')
plt.ylabel('Custo')
plt.legend()
plt.xticks(rotation=45)  # Rotacionar os rótulos dos meses para melhor legibilidade

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```

## Conclusão
Ao observar o gráfico, é evidente que houve um aumento considerável no custo de entrega em relação aos demais custos. O custo de produtos manteve-se estável durante os 9 meses, enquanto o custo de marketing apresentou uma diminuição ao longo do período. Em contraste, o custo de entrega segue uma tendência crescente mês a mês, conforme mostrado no gráfico.

Levando em consideração que o volume de vendas se manteve estável, é compreensível que o aumento gradual do custo de entrega (começando no 2º mês) impacta diretamente o lucro da empresa, comprometendo a rentabilidade.

## Explicação do Aumento no Custo de Entrega
O aumento no custo de entrega, que geralmente compreende os custos logísticos, pode ser explicado por várias variáveis:

**Preços dos Combustíveis**: O aumento nos preços dos combustíveis afeta diretamente a cadeia logística. Com preços mais altos para o combustível, o custo do frete para entrega dos produtos aumenta.

**Mudanças nos Custos Trabalhistas**: A elevação nos salários ou o pagamento de horas extras para equipes logísticas, contribui para o aumento dos custos de entrega.

**Mudanças Regulatórias**: Novas regulamentações ou impostos relacionados a serviços de transporte e entrega também podem aumentar o custo de entrega.

**Distribuição Geográfica**: O esforço para expandir as vendas em regiões geográficas remotas, menos acessíveis ou distantes dos centros de distribuição (CDs) e lojas pode aumentar significativamente o custo de frete. Por exemplo, as regiões Norte e Nordeste do Brasil, que somam mais de 63% do território nacional, representam um desafio considerável para a gestão dos custos de entrega devido às distâncias percorridas.

**Impacto no Negócio**
O aumento contínuo do custo de entrega, se não for controlado, pode reduzir as margens de lucro e afetar a rentabilidade da empresa. Recomenda-se estratégias para mitigar esses custos, como otimização logística, renegociação de contratos de transporte, e avaliação contínua dos impactos das variáveis mencionadas.

**Sugestões de Ações**
**Análise de Eficiência Logística**: Avaliar se há oportunidades para otimização na cadeia logística para reduzir custos.

**Contratos de Transporte**: Revisar e renegociar contratos com fornecedores de transporte para obter melhores condições.

**Monitoramento de Custos**: Implementar um sistema de monitoramento contínuo dos custos de entrega e das variáveis que afetam esses custos.

**Avaliação Regional**: Analisar a viabilidade de ajustar a estratégia de distribuição para regiões com maiores custos e identificar soluções mais econômicas.


## 3. Aprofundamento da Análise

Levando em consideração o aumento do custo de entrega, podemos entender melhor qual a correlação entre os custos, afim de capturar mais insights.



```python
correlacao = df.corr()
plot = sns.heatmap(correlacao, annot = True, fmt=".1f", linewidths=.6)
plot
```

## Mapa de  Correlação

O mapa térmico mostra os coeficientes de correlação entre cada par de variáveis, fornecendo insights sobre como os custos estão relacionados. 
Um coeficiente de correlação próximo de 1 indica uma forte correlação positiva, ou seja, à medida que uma variável aumenta a outra tende a aumentar também. 
Um coeficiente próximo de -1 indica uma forte correlação negativa, enquanto uma variável aumenta a outra diminui. 
Já um coeficiente em torno de 0 sugere não haver correlação linear entre as variáveis.

É possível observar esse nível de correlação quando analisamos a média.

```python
df_media_custos = df.groupby('categoria').mean()[['custo_entrega', 'custo_produtos','custo_marketing' ]]
df_media_custos
```

```python
# Ordenar a média do maior para o menor
df_media_custos = df_media_custos.sort_values(by=['custo_entrega', 'custo_produtos','custo_marketing'],
                                             ascending=False)
```

```python
df_media_custos
```

```python
#  Resetar index
df_media_custos = df_media_custos.reset_index()
df_media_custos
```

```python
# Formatar os números
df_media_custos['custo_marketing'] = df_media_custos['custo_marketing'].round(2)
df_media_custos
```

```python
df_media_custos['custo_produtos'] = df_media_custos['custo_produtos'].round(2)
df_media_custos
```

```python
df_media_custos['custo_entrega'] = df_media_custos['custo_entrega'].round(2)
df_media_custos
```
